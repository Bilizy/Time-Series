```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import dateutil

import warnings
warnings.filterwarnings("ignore")
```

```python
!tree ../. -L 2
```

```python
df = pd.read_csv('../data/example_air_passengers.csv', parse_dates=['ds'])
df['y_log'] = np.log(df.y)
df['y_log_dif'] = df.y_log.diff(1)
df.dropna(axis=0, inplace=True)

print('start date as: \t{}'.format(datetime.strftime(df.ds.min(), '%Y-%m-%d')))
print('end date as: \t{}'.format(datetime.strftime(df.ds.max(), '%Y-%m-%d')))
df
```

```python
length = 12
oneday = timedelta(days=1)
onemonth = dateutil.relativedelta.relativedelta(months=1)
predict_length = dateutil.relativedelta.relativedelta(months=length)
# onemonth = timedelta(days=predict_length)

date_start_predict = datetime.strftime(df.ds.max() + onemonth, '%Y-%m-%d')
date_end_predict = datetime.strftime(df.ds.max() + onemonth + predict_length, '%Y-%m-%d')
print('predict start date: {}'.format(date_start_predict))
print('predict end date: {}'.format(date_end_predict))
```

# Time Series Method
- AR (AutoRegressive Model)
- MA (Moving Average Model)

## Check

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import arma_order_select_ic
```

```python
def plot_acf_pacf(x, lags=20):
    fig = plt.figure(figsize=(12,4))
    ax1 = fig.add_subplot(121)
    plot_acf(x, ax=ax1, lags=lags)
    ax2 = fig.add_subplot(122)
    plot_pacf(x, ax=ax2, lags=lags)
#    return fig
```

```python
plot_acf_pacf(df.ds)
```

```python
plot_acf_pacf(df.y_log_dif)
```

```python
arma_order_select_ic(df.y_log_dif, max_ar=6, max_ma=6, ic=['aic', 'bic', 'hqic'])
# arma_order_select_ic(df.y, max_ar=5, max_ma=5, ic=['aic', 'bic', 'hqic'])
```

## AR Model
$$Y_t = \alpha Y_{t-1} + \varepsilon_i$$

```python
from statsmodels.tsa.api import ARMA
```

```python
model_ar = ARMA(df.y_log_dif, (1,0)).fit()
model_ma = ARMA(df.y_log_dif, (0,1)).fit()
model_arma = ARMA(df.y_log_dif, (1,1)).fit()
model_arma_ic = ARMA(df.y_log_dif, (6,5)).fit()
```

```python
plot_acf_pacf(model_ar.resid)
plot_acf_pacf(model_ma.resid)
plot_acf_pacf(model_arma.resid)
plot_acf_pacf(model_arma_ic.resid)
```

```python
pred_range = pd.date_range(start=date_start_predict, end=date_end_predict, freq='M')

# AR model
pred_ar = model_ar.predict(start=df.shape[0], end=df.shape[0]-1+length)
df_pred = pd.DataFrame({'ds': pred_range, 'yhat_ar_dif': pred_ar})

# MA model
pred_ma = model_ma.predict(start=df.shape[0], end=df.shape[0]-1+length)
df_pred['yhat_ma_dif'] = pred_ma

# ARMA model
pred_arma = model_arma.predict(start=df.shape[0], end=df.shape[0]-1+length)
df_pred['yhat_arma_dif'] = pred_arma

# ARMA IC model
pred_arma_ic = model_arma_ic.predict(start=df.shape[0], end=df.shape[0]-1+length)
df_pred['yhat_arma_ic_dif'] = pred_arma_ic

df_pred['ds'] = df_pred['ds'] + oneday - pd.DateOffset(months=1)
df_pred
```

```python
train_last_value = df.iloc[df.shape[0]-1, :].y_log
```

```python
pd.Series(train_last_value).append(df_pred.yhat_ar_dif).cumsum().values[1:]
```

```python
df_pred['yhat_ar'] = np.exp(pd.Series(train_last_value).append(df_pred.yhat_ar_dif).cumsum().values[1:])
df_pred['yhat_ma'] = np.exp(pd.Series(train_last_value).append(df_pred.yhat_ma_dif).cumsum().values[1:])
df_pred['yhat_arma'] = np.exp(pd.Series(train_last_value).append(df_pred.yhat_arma_dif).cumsum().values[1:])
df_pred['yhat_arma_ic'] = np.exp(pd.Series(train_last_value).append(df_pred.yhat_arma_ic_dif).cumsum().values[1:])
df_pred
```

```python
plt.figure(figsize=(18, 8))
plt.plot(df[df.ds>'1955'].ds, df[df.ds>'1955'].y_log_dif, label='actual')
plt.plot(df_pred.ds, df_pred.yhat_ar_dif, label='AR predict')
plt.plot(df_pred.ds, df_pred.yhat_ma_dif, label='MA predict')
plt.plot(df_pred.ds, df_pred.yhat_arma_dif, label='ARMA predict')
plt.plot(df_pred.ds, df_pred.yhat_arma_ic_dif, label='ARMA IC predict')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18, 8))
plt.plot(df[df.ds>'1955'].ds, df[df.ds>'1955'].y, label='actual')
plt.plot(df_pred.ds, df_pred.yhat_ar, label='AR predict')
plt.plot(df_pred.ds, df_pred.yhat_ma, label='MA predict')
plt.plot(df_pred.ds, df_pred.yhat_arma, label='ARMA predict')
plt.plot(df_pred.ds, df_pred.yhat_arma_ic, label='ARMA IC predict')
plt.legend()

plt.show()
```
